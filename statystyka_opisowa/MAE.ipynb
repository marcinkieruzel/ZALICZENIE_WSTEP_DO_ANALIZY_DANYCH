```python
import numpy as np
from sklearn.metrics import mean_absolute_error
import pandas as pd


#!!! Uwaga dane są sztucznie wygenerowane dla celów demonstracyjnych jak poniżej
data = [1,2,1,2]
predictions = [1,2,3,2]

mae = mean_absolute_error(data, predictions)
print(f"Mean Absolute Error (MAE): {mae}")

# 1 + 2 + 1 + 2 = 6 - data
# 1 + 2 + 3 + 2 = 8 - predictions
# |1-1| + |2-2| + |1-3| + |2-2| = 0 + 0 + 2 + 0 = 2
# MAE = 2 / 4 = 0.5
```
