```python
import numpy as np
from sklearn.metrics import mean_squared_error
import pandas as pd


#!!! Uwaga dane są sztucznie wygenerowane dla celów demonstracyjnych jak poniżej
data = [1,2,1,2]
predictions = [1,2,3,2]

mse = mean_squared_error(data, predictions)
print(f"Mean Squared Error (MSE): {mse}")

# Output:
# Mean Squared Error (MSE): 0.25
# 1 + 2 + 1 + 2 = 6 – data
# 1 + 2 + 3 + 2 = 8 – predictions
# Różnica: 2
# Kwadrat różnicy: 2^2 = 4
# Liczba próbek: 4
# Średnia z kwadratów różnic: 4 / 4 = 1.0
```
